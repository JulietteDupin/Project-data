```python
# Liste des colonnes représentant les maladies
disease_columns = [
    'SMOKING ', 'ALCOHOL', 'DM', 'HTN', 'CAD', 'PRIOR CMP', 'CKD', 'HB', 'TLC', 
    'PLATELETS', 'GLUCOSE', 'UREA', 'CREATININE', 'BNP', 'RAISED CARDIAC ENZYMES', 'EF', 
    'SEVERE ANAEMIA', 'ANAEMIA', 'STABLE ANGINA', 'ACS', 'STEMI', 'ATYPICAL CHEST PAIN', 
    'HEART FAILURE', 'HFREF', 'HFNEF', 'VALVULAR', 'CHB', 'SSS', 'AKI', 'CVA INFRACT', 
    'CVA BLEED', 'AF', 'VT', 'PSVT', 'CONGENITAL', 'UTI', 'NEURO CARDIOGENIC SYNCOPE', 
    'ORTHOSTATIC', 'INFECTIVE ENDOCARDITIS', 'DVT', 'CARDIOGENIC SHOCK', 'SHOCK', 
    'PULMONARY EMBOLISM', 'CHEST INFECTION'
]

# Convertir les valeurs dans les colonnes de maladies en entiers (0 ou 1)
df[disease_columns] = df[disease_columns].apply(pd.to_numeric, errors='coerce').fillna(0).astype(int)

# Calculer le nombre de patients par maladie (compter les valeurs égales à 1)
disease_counts = df[disease_columns].sum()

# Tracer le nombre de patients par maladie
plt.figure(figsize=(12,8))
disease_counts.plot(kind='bar', color='skyblue')
plt.title('Nombre de patients par maladie')
plt.xlabel('Maladies')
plt.ylabel('Nombre de patients')
plt.xticks(rotation=90)
plt.show()
```

```python
def get_season(month):
    """Retourne la saison basée sur le mois."""
    if month in [12, 1, 2]:  # Hiver
        return 'Hiver'
    elif month in [3, 4, 5]:  # Printemps
        return 'Printemps'
    elif month in [6, 7, 8]:  # Été
        return 'Été'
    else:  # Automne
        return 'Automne'

# 1. Convertir la colonne 'month year' en datetime en format 'mmm-yy' (ex: Apr-17)
df['month year'] = pd.to_datetime(df['month year'], format='%b-%y')

# 2. Extraire le mois et l'année de 'month year'
df['month'] = df['month year'].dt.month
df['year'] = df['month year'].dt.year

# 3. Appliquer la fonction pour obtenir la saison à partir du mois
df['Saison'] = df['month'].apply(get_season)

# 4. Convertir les colonnes de maladies en valeurs numériques (0 ou 1)
df[disease_columns] = df[disease_columns].apply(pd.to_numeric, errors='coerce').fillna(0).astype(int)

# 5. Calculer le nombre de patients malades par maladie et par saison
disease_season_counts = df.groupby('month year')[disease_columns].sum()

# 6. Visualiser le nombre de patients malades par maladie et par saison
disease_season_counts.plot(kind='bar', figsize=(14,8), stacked=True)
plt.title('Nombre de patients malades par maladie et par saison')
plt.xlabel('month year')
plt.ylabel('Nombre de patients malades')
plt.xticks(rotation=0)
plt.legend(title="Maladies", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# 📌 Charger les données
df = pd.read_csv('Backend/datas/HDHI Admission Data.csv', encoding='utf-8')

# 📌 Vérifier les colonnes
df.columns = df.columns.str.strip()  # Supprime les espaces autour des noms de colonnes
print("Colonnes disponibles:", df.columns)

# Sélectionner les colonnes nécessaires
selected_columns = ["month year", "HB", "TLC", "PLATELETS", "GLUCOSE", "UREA", "CREATININE", "BNP", "EF"]
df = df[selected_columns]

# 📌 Convertir 'month year' en datetime et utiliser comme index
df["month year"] = pd.to_datetime(df["month year"], format="%b-%y", errors='coerce')

# Vérifier si certaines dates n'ont pas pu être converties
if df["month year"].isna().sum() > 0:
    print("⚠️ Certaines dates n'ont pas pu être converties. Vérifie le format dans le CSV.")
    print(df[df["month year"].isna()])

df.set_index("month year", inplace=True)

# 📌 Assurer que toutes les colonnes de maladies sont numériques et gérer les NaN
for col in df.columns:
    # Convertir en numérique (forces les erreurs NaN)
    df[col] = pd.to_numeric(df[col], errors='coerce')
    # Gérer les NaN (remplacer par 0, ou choisir une autre méthode comme la moyenne)
    df[col].fillna(0, inplace=True)

# 📌 Agréger les données par mois (utiliser "ME" au lieu de "M")
df_monthly = df.resample("ME").sum()

# 📌 Entraînement du modèle ARIMA et prévision
predictions = {}

for disease in df_monthly.columns:
    print(f"🔎 Entraînement du modèle ARIMA pour {disease}...")

    if df_monthly[disease].nunique() < 5:
        print(f"⚠️ Trop peu de données pour {disease}, skipping...")
        continue

    # 📌 Vérifier si les données sont bien numériques
    print(f"Type de données pour {disease}: {df_monthly[disease].dtype}")
    
    try:
        # Entraînement du modèle ARIMA
        model = ARIMA(df_monthly[disease], order=(2, 1, 2))  # (p, d, q) ajustables
        model_fit = model.fit()

        # Prédire le mois suivant
        forecast = model_fit.forecast(steps=1)
        predictions[disease] = forecast.iloc[0]   # Remplacer par forecast.iloc[0] si tu veux éviter la dépréciation

        # 📊 Affichage des prévisions
        plt.figure(figsize=(10, 4))
        plt.plot(df_monthly[disease], label="Données réelles")
        plt.axvline(df_monthly.index[-1], color='r', linestyle='--', label="Début de la prévision")
        plt.scatter(df_monthly.index[-1] + pd.DateOffset(months=1), forecast, color='red', label="Prévision")
        plt.title(f"Prévision ARIMA pour {disease}")
        plt.legend()
        plt.show()

    except Exception as e:
        print(f"❌ Erreur pour {disease} : {e}")
        continue

# 📊 Afficher les prévisions
print("\n📊 Prédictions du nombre de patients pour le mois prochain :")
for disease, value in predictions.items():
    print(f"🩺 {disease} : {value} patients")
```
